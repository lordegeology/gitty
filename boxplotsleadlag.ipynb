```python
x = df_brazil['variable'].to_numpy()
```

```python
y = df_brazil['value'].to_numpy()
```

```python
z = df_brazil['LS'].to_numpy()
z
```

```python
x
```

```python
chunks = list()
k = 0
for j in range(10):
  
  for i in range(0, len(y), 768):
    chunks.append(y[i+k:768+k])
  k = k+768

chunks = [x for x in chunks if x != []]
```

```python
len(chunks)
```

```python
chunks
```

```python
fig = plt.figure(figsize =(10, 10),facecolor=(1, 1, 1))
 
ax = fig.add_axes([0, 0, 1, 1])
ax.grid(color='white', axis='y', linestyle='-', linewidth=0.25, alpha=0.5)

# Set plot title
ax.set_title('Boxplots Lead Lag Brazil')
bp = ax.boxplot(chunks, patch_artist = True, 
                notch ='True', vert = 1)

   
# changing color and linewidth of 
# caps 
for cap in bp['caps']: 
    cap.set(color ='#8B008B', 
            linewidth = 2) 
   
# changing color and linewidth of 
# medians 
for median in bp['medians']: 
    median.set(color ='red', 
               linewidth = 3) 
 
labels = [item.get_text() for item in ax.get_xticklabels()]
labels = ["Lag.10", "Lag.9","Lag.8","Lag.7","Lag.6","Lag.5","Lag.4","Lag.3","Lag.2","Lag.1"]
ax.set_xticklabels(labels)
plt.show()
```

```python
outlist = []
d = {}
for i in range(10):
  # finding the 1st quartile
  q1 = np.quantile(chunks[i], 0.25)
 
  # finding the 3rd quartile
  q3 = np.quantile(chunks[i], 0.75)
  med = np.median(chunks[i])
 
  # finding the iqr region
  iqr = q3-q1
 
  # finding upper and lower whiskers
  upper_bound = q3+(1.5*iqr)
  lower_bound = q1-(1.5*iqr)
  print(iqr, upper_bound, lower_bound)

  outliers = chunks[i][(chunks[i] <= lower_bound) | (chunks[i] >= upper_bound)]
  print('The following are the outliers in the boxplot:{}'.format(outliers))
  for j in range(len(outliers)):
    
    for k in range(len(y)):
      if outliers[j] == y[k]:
        outlist.append(z[k])
  d[i] = outlist
  
  print(outlist)
  
  
  
  outlist = []
  

```
